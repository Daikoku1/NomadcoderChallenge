```python

def add_to_dict(*args):
    li = list(args)
    if type(li[0]) != dict:
        print(f'You need to send a dictionary. You sent: {type(li[0])}')
        return 'First argument should be a dict.'
    elif len(li) < 3:
        print(f'You need to send a word and a definition')
        return 'Definition is required.'
    elif len(li) > 3:
        return 'Too many arguments'

    if li[1] in li[0].keys():
        print(f"{li[1]} is already on the dictionary. Won't add.")
        return  f'{li[1]} is already on the dict'
    print(f'{li[1]} has been added')
    li[0][li[1]] = li[2]


def get_from_dict(*args):
    li = list(args)
    if type(li[0]) != dict:
        print(f'You need to send a dictionary. You sent: {type(li[0])}')
        return 'First argument should be a dict.'
    if len(li) < 2:
        print(f'You need to send a word to search for.')
        return 'Word to search from is required.'
    if len(li) > 2:
        return 'Too many arguments'
    try:
        print(li[0][li[1]])
        return li[0][li[1]]
    except:
        print(f"{li[1]} was not found in this dict.")
        return 'Word is not found.'


def update_word(*args):
    li = list(args)
    if type(li[0]) != dict:
        print(f'You need to send a dictionary. You sent: {type(li[0])}')
        return 'First argument should be a dict.'
    if len(li) < 3:
        print(f'You need to send a word and a definition to update.')
        return 'Word and definiton are required.'
    if len(li) > 3:
        return 'Too many arguments'
    
    if li[1] in li[0].keys():
        print(f"{li[1]} has been updated to: {li[2]}")
        li[0][li[1]] = li[2]
    else:
        print(f"{li[1]} is not on the dict. Cant't update non-existing word.")
        return 'Word not found.'


def delete_from_dict(*args):
    li = list(args)
    if type(li[0]) != dict:
        print(f'You need to send a dictionary. You sent: {type(li[0])}')
        return 'First argument should be a dict.'
    if len(li) < 2:
        print(f'You need to specify a word to delete.')
        return 'Word to delete is required.'
    if len(li) > 2:
        return 'Too many arguments'
    
    if li[1] in li[0].keys():
        print(f"{li[1]} has been deleted.")
        del li[0][li[1]]
    else:
        print(f"{li[1]} is not in this dict. Won't delete.")
        return 'Word not found.'
```
